```python
import chardet
import numpy as np
import pandas as pd
import datetime
import codecs

import os
import glob

import progressbar
from time import sleep

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
moraine = pd.read_csv( "../data/Site-Data/moraine_agg.csv")
mather = pd.read_csv( "../data/Site-Data/mather_agg.csv")
big_meadow = pd.read_csv( "../data/Site-Data/bigmeadow_agg.csv")
upperpines = pd.read_csv( "../data/Site-Data/upperpines_agg.csv")
elkmont = pd.read_csv( "../data/Site-Data/elkmont_agg.csv")
```

```python
moraine.index = pd.DatetimeIndex(moraine['Date'])
mather.index = pd.DatetimeIndex(mather['Date'])
big_meadow.index = pd.DatetimeIndex(big_meadow['Date'])
upperpines.index = pd.DatetimeIndex(upperpines['Date'])
elkmont.index = pd.DatetimeIndex(elkmont['Date'])
```

```python
fig, axes = plt.subplots(2, 3, figsize=(20, 12))

axes[0][0].plot(moraine.index, moraine['PercentBooked'], alpha=0.75)
axes[0][0].set_title('Moraine')

axes[0][1].plot(mather.index, mather['PercentBooked'], alpha=0.75)
axes[0][1].set_title('Mather')

axes[0][2].plot(big_meadow.index, big_meadow['PercentBooked'], alpha=0.75)
axes[0][2].set_title('Big Meadows')

axes[1][0].plot(upperpines.index, upperpines['PercentBooked'], alpha=0.75)
axes[1][0].set_title('Upper Pines')

axes[1][1].plot(elkmont.index, elkmont['PercentBooked'], alpha=0.75)
axes[1][1].set_title('Elkmont')
```
